```python
A = 4.0478 * 0.4991 / 0.5
volumes = []
energies = []
for i in tqdm(np.arange(-5, 5, 0.5)):
    if i == 0:
        continue
    with open(r"HW/al2.in", 'w') as f:
        text = '''
&control
    calculation = 'scf'
    prefix = 'cubic'
    pseudo_dir = '/home/vadim/Документы/Quantum/q-e-qe-7.0/sssp'
    outdir = '/home/vadim/Документы/Quantum/q-e-qe-7.0/out'
/
&system
    ibrav = 2
    A = {}
    nat = 1
    ntyp = 1
    ecutwfc = 60
    ecutrho = 480
    occupations = 'smearing'
    smearing = 'cold'
    degauss = 0.02
/
&electrons
    conv_thr=1.d-8
/
&ions
/
&cell
/
ATOMIC_SPECIES
    Al 26.981 Al.pbe-n-kjpaw_psl.1.0.0.UPF
ATOMIC_POSITIONS crystal
    Al 0.0 0.0 0.0
K_POINTS automatic
    20 20 20 0 0 0
         '''.format(A * (1 + i/100))
        f.write(text)
        f.close()
        res = !mpirun -n 4 bin/pw.x -in HW/al2.in
        res = ''.join(res)
        #print(res)
        volumes.append((A * (1 + i/100)) ** 3)
        energy = re.findall(r"!\s+total energy\s+=\s+-[\d\.]+", res)[0]
        energy = float(re.findall(r"-[\d\.]+", energy)[0])
        energies.append(energy)
```

```python
4.0478 * 0.4991 / 0.5
```

```python
def new_eos(v, e0, b0, v0, d):
    temp = ((v0/v)**(2/3) - 1)
    val = e0 + 9*b0*v0/16*(temp ** 3 * d + 
                           temp ** 2 * (6 - 4*(v0/v)**(2/3)))
    return val
```

```python
bounds=(
    (-50, -np.inf, 63, 1),
    (-30, np.inf, 67, 1000)
)
fit, cov = curve_fit(new_eos, volumes, energies, bounds=bounds,
                    method='trf')
```
